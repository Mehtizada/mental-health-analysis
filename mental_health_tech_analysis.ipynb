```python
df = pd.read_csv('/survey.csv')
```

```python
print("İlk 5 sətir:")
print(df.head())
print("\nVerilənlərin ölçüsü:", df.shape)
print("\nBoş olan sütunlar:")
print(df.isnull().sum())
```

```python
print("\nCins bölgüsü:")
print(df['Gender'].value_counts())
```

```python
print("\nMinimum yaş:", df['Age'].min())
print("Maksimum yaş:", df['Age'].max())
```

```python
df = df[(df['Age'] >= 15) & (df['Age'] <= 65)]
```

```python
sns.countplot(data=df, x='work_interfere', hue='mental_health_consequence')
plt.title('Mental health consequences vs. work interference')
plt.xticks(rotation=30)
plt.show()
```

```python
sns.countplot(data=df, x='Gender', hue='benefits')
plt.title('Gender vs. Access to Mental Health Benefits')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='care_options', hue='Gender')
plt.title('Comfort in discussing mental health at workplace')
plt.xticks(rotation=45)
plt.show()
```
